```python
import os
import json

def format_header():
    return 'date,seqnum,qty,unitprice,product,total,savings\n'

def format_line_item(item, date, seqnum):
    qty = item['qty'] if 'qty' in item else ''
    unitprice = item['unitprice'] if 'unitprice' in item else ''
    product = item['product']
    total = item['total']
    savings = item['savings'] if 'savings' in item else ''
    return f'{date},{seqnum},{qty},{unitprice},{product},{total},{savings}'

def format_line_items(items, date, seqnum):
    buf = ''
    for item in items:
        s = format_line_item(item, date, seqnum)
        buf = buf + s + '\n'
    return buf

def write_header_file(base_path):
    path = f'{base_path}/CSV/0_header.csv'
    f = open(path, 'w')
    f.write(format_header())
    f.close()

def convert_json_to_csv(base_path, receipt_name):
    json_path = f'{base_path}/JSON'
    csv_path = f'{base_path}/CSV'
    path = f'{json_path}/{receipt_name}.json'
    f = open(path, 'r')
    j = json.load(f)
    f.close()
    header = j['header']
    receipt_date = header['receipt_date']
    receipt_seqnum = header['receipt_seqnum']
    path = f'{csv_path}/{receipt_name}.csv'
    line_items = j['body']['line_items']
    f = open(path, 'w')
    csv = format_line_items(line_items, receipt_date, receipt_seqnum)
    f.write(csv)
    f.close()
```

```python
chunk = 0
chunksize = 500
problem_receipts = [
#    '2020_04_19_01',
                   ]

base_path = '/Users/ericmelz/Desktop/Ralphs/Receipts'
json_path = f'{base_path}/json'

l = os.listdir(json_path)
all_receipt_names = sorted([f.split('.')[0] for f in l])

start_offset = chunksize * chunk
end_offset = start_offset + chunksize
receipt_names = all_receipt_names[start_offset:end_offset]

write_header_file(base_path)

for receipt_name in receipt_names:
    print(f'Converting {receipt_name}....', end='')
    if receipt_name in problem_receipts:
        print('SKIPPING!')
        continue
    print()
    convert_json_to_csv(base_path, receipt_name)
```
